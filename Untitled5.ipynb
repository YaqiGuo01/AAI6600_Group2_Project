```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCA to 2D
pca = PCA(n_components=2, random_state=42)
coords = pca.fit_transform(X)

df["pc1"] = coords[:, 0]
df["pc2"] = coords[:, 1]

# Scatter plot
plt.figure(figsize=(10, 7))
plt.scatter(df["pc1"], df["pc2"], alpha=0.7, s=50, color="steelblue")
plt.title("Healthcare Questions: Gemini Embeddings + PCA(2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

inertias = []
k_values = [4, 5, 6, 7, 8, 9, 10, 11, 12]

for k in k_values:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertias, marker='o', color='steelblue')
plt.title("Elbow Method — Choosing k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print("Chosen k = 7")
print("""Justification: The elbow curve shows a noticeable change in slope around k=7,
where the rate of inertia decrease begins to slow down.
This also aligns well with the natural intent categories in the dataset
(appointments, medications, lab results, billing, telehealth, records, and urgent care).""")
```

```python
k = 7
km = KMeans(n_clusters=k, random_state=42, n_init=10)
df["cluster"] = km.fit_predict(X)

# Cluster sizes descending
sizes = df["cluster"].value_counts().sort_values(ascending=False)
print("Cluster sizes (descending):")
print(sizes.to_string())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

print("=== Representative Questions per Cluster ===\n")
for cluster_id in range(k):
    mask = df["cluster"] == cluster_id
    cluster_indices = df[mask].index.tolist()
    cluster_embeddings = X[cluster_indices]

    centroid = km.cluster_centers_[cluster_id].reshape(1, -1)
    sims = cosine_similarity(centroid, cluster_embeddings)[0]

    top5_local = np.argsort(sims)[::-1][:5]
    top5_global = [cluster_indices[i] for i in top5_local]

    print(f"--- Cluster {cluster_id} ---")
    for idx in top5_global:
        print(f"  {df['question'][idx]}")
    print()
```

```python
cluster_intent = {
    0: "Medication Side Effects",
    1: "Lab Results Access",
    2: "Medical Records Transfer",
    3: "Prescription Refills",
    4: "Appointment Scheduling",
    5: "Profile & Contact Update",
    6: "Appointment Changes",
}

df["intent"] = df["cluster"].map(cluster_intent)

print("Cluster → Intent Mapping:")
for cluster_id, name in cluster_intent.items():
    print(f"  Cluster {cluster_id}: {name}")

print("\n3 Example Questions per Intent:")
for cluster_id, intent_name in cluster_intent.items():
    subset = df[df["cluster"] == cluster_id]["question"].head(3)
    print(f"\n[{intent_name}]")
    for q in subset:
        print(f"  - {q}")
```

```python
# Top 2 largest clusters (from earlier: cluster 4=32, cluster 2=16)
top2_clusters = df["cluster"].value_counts().nlargest(2).index.tolist()
print(f"Top 2 clusters: {top2_clusters}")
print(f"  Cluster {top2_clusters[0]}: {cluster_intent[top2_clusters[0]]}")
print(f"  Cluster {top2_clusters[1]}: {cluster_intent[top2_clusters[1]]}")

# Select 3 canonical FAQs from each (closest to centroid)
canonical_faqs = []
canonical_intents = []

for cluster_id in top2_clusters:
    mask = df["cluster"] == cluster_id
    cluster_indices = df[mask].index.tolist()
    cluster_embeddings = X[cluster_indices]
    centroid = km.cluster_centers_[cluster_id].reshape(1, -1)
    sims = cosine_similarity(centroid, cluster_embeddings)[0]
    top3_local = np.argsort(sims)[::-1][:3]
    top3_global = [cluster_indices[i] for i in top3_local]
    for idx in top3_global:
        canonical_faqs.append(df["question"][idx])
        canonical_intents.append(cluster_intent[cluster_id])

print("\nCanonical FAQs selected:")
for faq, intent in zip(canonical_faqs, canonical_intents):
    print(f"  [{intent}] {faq}")
```

```python
faq_result = client.models.embed_content(
    model="models/gemini-embedding-001",
    contents=canonical_faqs,
)

faq_embeddings = np.array([e.values for e in faq_result.embeddings])
print(f"FAQ embeddings shape: {faq_embeddings.shape}")
```

```python
THRESHOLD = 0.75

def chatbot(query):
    # 1. Embed query
    q_result = client.models.embed_content(
        model="models/gemini-embedding-001",
        contents=[query],
    )
    q_vec = np.array([e.values for e in q_result.embeddings])

    # 2. Cosine similarity to all canonical FAQs
    sims = cosine_similarity(q_vec, faq_embeddings)[0]
    best_idx = np.argmax(sims)
    best_sim = sims[best_idx]
    best_faq = canonical_faqs[best_idx]
    best_intent = canonical_intents[best_idx]

    # 3. Decision
    print(f"\n{'='*55}")
    print(f"Query:            {query}")
    print(f"Predicted Intent: {best_intent}")
    print(f"Matched FAQ:      {best_faq}")
    print(f"Similarity:       {best_sim:.4f}")

    if best_sim >= THRESHOLD:
        print(f"Decision:         ANSWER")
        print(f"\nAnswer: Based on your question, please log in to the patient portal or contact our clinic for assistance.")
    else:
        print(f"Decision:         REFUSE")
        print(f"\nI'm sorry, I can only answer questions related to Appointment Scheduling and Medical Records Transfer.")

    print(f"\nNot medical advice. If this is an emergency, call 911.")
    print(f"{'='*55}")
```

```python
test_queries = [
    "I want to book a visit for next week.",
    "How do I get my lab results?",
    "Can you explain my MRI and diagnose me?",
    "What time does the clinic open?",
]

for q in test_queries:
    chatbot(q)
```
